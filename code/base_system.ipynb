```python
import random

# Define the environment
class IsolatedEnvironment(cr.Environment):
    def __init__(self):
        super().__init__()
        self.resources = 100
        self.infrastructure_health = 100
        self.health_status = 100
        self.conflict_level = 0

    def get_state(self):
        return {
            'resources': self.resources,
            'infrastructure_health': self.infrastructure_health,
            'health_status': self.health_status,
            'conflict_level': self.conflict_level,
        }

    def update_state(self, updates):
        for key, value in updates.items():
            setattr(self, key, value)

# Define the Resource Management Agent
class ResourceManagementAgent(cr.Agent):
    def act(self, state):
        if state['resources'] < 50:
            # Simulate resource allocation optimization
            new_resources = state['resources'] + random.randint(10, 20)
            return {'resources': new_resources}
        return {}

# Define the Maintenance Agent
class MaintenanceAgent(cr.Agent):
    def act(self, state):
        if state['infrastructure_health'] < 50:
            # Simulate preventive maintenance
            new_infrastructure_health = state['infrastructure_health'] + random.randint(10, 20)
            return {'infrastructure_health': new_infrastructure_health}
        return {}

# Define the Health Monitoring Agent
class HealthMonitoringAgent(cr.Agent):
    def act(self, state):
        if state['health_status'] < 50:
            # Simulate health intervention
            new_health_status = state['health_status'] + random.randint(10, 20)
            return {'health_status': new_health_status}
        return {}

# Define the Conflict Management Agent
class ConflictManagementAgent(cr.Agent):
    def act(self, state):
        if state['conflict_level'] > 50:
            # Simulate conflict resolution
            new_conflict_level = state['conflict_level'] - random.randint(10, 20)
            return {'conflict_level': new_conflict_level}
        return {}

# Initialize the environment and agents
env = IsolatedEnvironment()
agents = [
    ResourceManagementAgent(),
    MaintenanceAgent(),
    HealthMonitoringAgent(),
    ConflictManagementAgent()
]

# Simulate the system
for _ in range(10):  # Run for 10 cycles
    state = env.get_state()
    print(f"State before actions: {state}")

    updates = {}
    for agent in agents:
        updates.update(agent.act(state))
    
    env.update_state(updates)
    state = env.get_state()
    print(f"State after actions: {state}")

    print("-" * 50)
```
